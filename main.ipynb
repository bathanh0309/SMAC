```python
# ================= CELL 2: SETUP SERVER & BLYNK =================
import requests
# Cấu hình Server
SERVER_URL = "http://192.168.1.100:8000"
BLYNK_TOKEN = "cE6rY7CdzQ5IvWtj2QhbKzhdNTV-F1tq"
BLYNK_VPIN = "V1"
def setup_blynk():
    print("Đang cập nhật Blynk với server cá nhân...")
    print("="*50)
    print(f"Server URL: {SERVER_URL}")
    print(f"Link Video: {SERVER_URL}/video")
    
    # Cập nhật Blynk
    blynk_update_url = "https://blynk.cloud/external/api/update"
    params = {
        "token": BLYNK_TOKEN,
        BLYNK_VPIN: f"{SERVER_URL}/video"
    }
    
    try:
        requests.get(blynk_update_url, params=params, timeout=5)
        print(f"Đã cập nhật Widget Video ({BLYNK_VPIN}) trên Blynk!")
    except Exception as e:
        print(f"Lỗi cập nhật Blynk: {e}")
        
    print("="*50)
    print("Setup hoàn tất! Chạy CELL 3 để bắt đầu phát hiện.")
    return SERVER_URL
# Chạy setup
setup_blynk()
```
